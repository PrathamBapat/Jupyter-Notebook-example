# Scipy.Optimize.Curve_fit in the IPython Notebook Example
      In this example we study how to use scipy.optimize

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

## h5py

   This package is necessary to extract the data from the hdf5 file
   
## scipy.optimize
   
   Use non-linear least squares to fit a function, f, to data.

```python
def func(x, a, b, c):
   return a * np.exp(-b * x) + c
```

As per the documentation lets consider the most basic of the functions.



```python
f = h5py.File("data.h5", 'r')
variable_path = f['AwakeEventData']['XMPP-STREAK']['StreakImage'] # Reading the data giving the required path 
xdata = variable_path['streakImageTimeValues']
```

Let our x-data be the random 512 values extracted from a dataset.

```python
ydata = np.linspace(0, 4, 512)
```

Let our y-data be a equally divided 512 samples in between [0,4]


## Let's plot the data using x and y.

```python
plt.plot(xdata, ydata, 'g--')
plt.show()
```

## Fit for the parameters a, b, c of the function func:
     Let's plot using the linear array of popt

```python
popt, pcov = curve_fit(func, xdata, ydata)
plt.plot(xdata, func(xdata, *popt), 'r-')
plt.show()
```
